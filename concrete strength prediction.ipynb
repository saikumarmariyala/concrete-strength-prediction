```python
from scipy import stats

lower_bound =df['cement'].quantile(q=0.25)
upper_bound =df['cement'].quantile(q=0.75)

print('1st quartile (Q1) is:',lower_bound)
print('3rd quartile (Q3) is:',upper_bound)
print('Interquartile range (IQR) is',stats.iqr(df['cement']))
```

```python
min_outliers = lower_bound-1.5*(upper_bound-lower_bound)
max_outliers = upper_bound+1.5*(upper_bound-lower_bound)
print('minimum outlier value in cement is:',min_outliers)
print('maximun outlier value in cement is:',max_outliers)
```

## replacing the outliers by median

```python
for cols in df.columns[:-1]:
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)
    iqr = Q3 - Q1
    
    low = Q1 - 1.5*iqr
    high= Q1 + 1.5*iqr
    df.loc[(df[cols]<low) | (df[cols]>high),cols] = df[cols].median()
    
```
